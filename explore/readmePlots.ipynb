```plaintext
import os
import sys

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

sys.path.append('/Users/John/Documents/allProjects/genericfunctions')

from FBB_points import *
seedA = 23
np.random.seed(seedA)
```

```plaintext
x1 = np.random.normal(10, 1, 100)
x2 = np.random.normal(30, 1, 100)
x = np.append(x1,x2)
med = np.median(x)
```

```plaintext
fig, ax = plt.subplots(figsize = (5,3))
fig = sns.kdeplot(x, ax = ax, label = 'Projection Values')
fig = plt.axvline(med, color = 'g', linestyle = '--', label ='Median')
ax.set_ylabel("Density")
ax.set_xlabel("Projection Outcomes")
ax.set_title("Example Bimodal Distribution")
ax.set_yticklabels(labels = [])
ax.set_xticklabels(labels = [])
ax.set_yticks([0,0.09])
ax.legend(loc = 'upper left')

#(visible = False)
ax.tick_params(left = False, bottom = False)
#ax.tick_params(axis='both', which='both', length=0)
plt.savefig(vars['data_clean']+'BiModal.png', format='png')
```
